# PyCity Schools Analysis

The PyCity Schools project analyzed 15 high schools across one district.  Raw data was composed of two csv files, â€˜schools_complete.csvâ€™ and â€˜students_complete.csvâ€™.  Schools_complete.csv comprised school-specific information, including school type, student population size and budget.  Accordingly,  students_complete.csv comprised student information such as grade, gender, high school, and reading & math scores.  The two files were combined, and the aggregated data was stratified and organized into visually compelling tables that illustrated data relationships, leading to â€˜potentially actionableâ€™ insights and decisions on the part of district-wide leaders and legislators.

First and foremost, charter schools outperform their traditional public school counterparts across all four grades in math and reading curricula.  Our analysis shows the top five highest performing schools (in both math and reading) are all charter, and no charter school is among the five lowest performing.  Could it be that charter schools having the freedom to tailor their own curriculum, intrinsically requires that they invest more time and consideration in identifying the needs of their students and the factors that most contribute to their academic success. Additionally, this freedom likely comes with the power to efficiently and creatively implement changes in quick turn-around time, all with the intention of improving student performance outcomes.

Furthermore, larger schools do not fare as well as their smaller counterparts.  The fact that in our data, all non-charter, i.e., â€˜districtâ€™, schools are categorized as â€˜largeâ€™, compared to only one â€˜large-sizedâ€™ charter does indeed impede the hasty conclusion that â€˜big is badâ€™.  As well, one â€˜large-sizedâ€™, a charter, Wilson High School, is among the five highest performing schools.  And further, the fact that the majority of  higher performing schools are categorized as â€˜mediumâ€™, reveals that, likewise, â€˜smallâ€™ does not necessarily mean â€˜betterâ€™.  Note, only one â€˜smallâ€™ school is among the top performers. 

And yet, these observations do not necessarily throw out the possibility that â€˜largeâ€™ institutions may have challenges impacting student success that are directly related to their size.  For instance, consider larger teacher-to-student ratios, or in other words, larger class sizes.  Larger overall staff-to-student ratios, which would include support mentors such as coaches, counselors, and safety officers, may also be a contributing factor.  Simply put, the impact of larger ratios leading to students having less one-on-one time with adult mentors invested in their success cannot be overlooked
In addition, the highest â€˜per student capitaâ€™ spending is found among the larger schools.  Clearly, it costs more to run a larger institution, but it is the â€˜per-capitaâ€™ funding which stands out here.  In fact, four of the five top â€˜per-capitaâ€™ spenders were among the lowest performing institutions. This may indeed reinforce the fact that â€˜money spentâ€™ cannot make up for the lack of one-on-one time with support staff, be it teachers or other.

In all, a sample size of 15 schools within one â€˜un-namedâ€™ district, and only a handful of analytic variables, is likely not enough to make wide-sweeping conclusions with regard to public education as a whole.  Additional school district socioeconomic and demographic information would be greatly helpful.  As well, bringing in data from surrounding districts, and perhaps that from other states, may lead to comparisons which either reinforce or dispel the insights proposed above, all of which could provide a cornucopia of impactful factors manifest in much needed decision-making wisdom ðŸ˜Š.






```python
import warnings
warnings.filterwarnings('ignore')


# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
students_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load, encoding="UTF-8")
student_data_df = pd.read_csv(students_data_to_load, encoding="UTF-8")

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# QA CHECK: Retrieve complete list of variable names
school_data_complete_df.columns
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete_df["School ID"].value_counts())
school_count
```

```python
# Calculate the total budget
total_budget = sum(school_data_complete_df["budget"].unique())
total_budget
```

```python
# Calculate the total number of students
student_count = len(school_data_complete_df["Student ID"].value_counts())
student_count
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete_df[
    (school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate

# QC CHECK: below should = 25528
#passing_math_reading_count
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame

district_summary_df = pd.DataFrame({
    "Total Schools" : school_count,
    "Total Students" : student_count,
    "Total Budget": total_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Students Passing Math": passing_math_percentage,
    "% Students Passing Reading": passing_reading_percentage,
    "% Students Overall Passing": overall_passing_rate
},index=[" "])

district_summary_df

# Formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary_df
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data_df.set_index("school_name")["type"]
school_types
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete_df["school_name"].value_counts()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
school_budget=school_data_complete_df.groupby("school_name").max()["budget"]
school_budget

student_budget=school_budget/per_school_counts
student_budget
```

```python
# Calculate the average test scores per school
per_school_math_avg = school_data_complete_df.groupby("school_name").mean()["math_score"]
per_school_reading_avg = school_data_complete_df.groupby("school_name").mean()["reading_score"]
```

```python
# Calculate the number of students per school with math scores of 70 or higher
per_school_passing_math = school_data_complete_df.loc[(school_data_complete_df["math_score"] >= 70)].groupby("school_name").count()["Student ID"]
per_school_passing_math

# QC CHECK - Sum of values above should match below (total number of students who passed math - 29370)
# passing_math_count = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)].count()["Student ID"]
# passing_math_count
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
per_school_passing_reading = school_data_complete_df.loc[(school_data_complete_df["reading_score"] >= 70)].groupby("school_name").count()["Student ID"]
per_school_passing_reading

# QC CHECK - Sum of values above should match below (total number of students who passed reading - 33610)
# passing_reading_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)].count()["Student ID"]
# passing_reading_count 
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete_df[
    (school_data_complete_df["reading_score"] >= 70) & (school_data_complete_df["math_score"] >= 70)
]

school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math_rate = per_school_passing_math / per_school_counts * 100
per_school_passing_reading_rate = per_school_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary_df = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": school_budget,
    "Per Student Budget": student_budget,
    "Average Math Score": per_school_math_avg,
    "Average Reading Score": per_school_reading_avg,
    "% Passing Math": per_school_passing_math_rate,
    "% Passing Reading": per_school_passing_reading_rate,
    "% Overall Passing": overall_passing_rate  
})

# Preserve an unformatted copy that does not have the formatting you will employ below (i.e. formatting turns the column into a string)
# Note, we will not be able to do calculations on strings (relevant to 'Scores by School Spending' section below).
per_school_summary_unfmt_df=per_school_summary_df.copy()

# Formatting
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary_df
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "9th")]
tenth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "10th")]
eleventh_graders = school_data_complete_df[(school_data_complete_df["grade"] == "11th")]
twelfth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_grade_math_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_grade_math_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_grade_math_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
    "9th Grade Math Scores" : ninth_grade_math_scores,
    "10th Grade Math Scores" : tenth_grade_math_scores,
    "11th Grade Math Scores" : eleventh_grade_math_scores,
    "12th Grade Math Scores" : twelfth_grade_math_scores 
})

math_scores_by_grade

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Reading Score by Grade 

```python
# Note, the data has been separated by grade in the previous step (so, no need to repeat the code -- yayyyy!!!!)

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading_scores = ninth_graders.groupby(["school_name"]).mean()["reading_score"]
tenth_grade_reading_scores = tenth_graders.groupby(["school_name"]).mean()["reading_score"]
eleventh_grade_reading_scores = eleventh_graders.groupby(["school_name"]).mean()["reading_score"]
twelfth_grade_reading_scores = twelfth_graders.groupby(["school_name"]).mean()["reading_score"]

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({
    "9th Grade Reading Scores" : ninth_grade_reading_scores,
    "10th Grade Reading Scores" : tenth_grade_reading_scores,
    "11th Grade Reading Scores" : eleventh_grade_reading_scores,
    "12th Grade Reading Scores" : twelfth_grade_reading_scores 
})

reading_scores_by_grade

# Minor data wrangling
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
bin_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_budget_bins_df = per_school_summary_unfmt_df.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_budget_bins_df["Spending Ranges (Per Student)"]=pd.cut(school_budget_bins_df["Per Student Budget"], spending_bins, labels=bin_labels)
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_budget_bins_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_budget_bins_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_budget_bins_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_budget_bins_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_budget_bins_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
school_summary_budget_bins_df = pd.DataFrame({
    "Average Math Score": spending_math_scores, 
    "Average Math Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

# Display results
school_summary_budget_bins_df
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_size_bins_df = per_school_summary_unfmt_df.copy()
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
school_size_bins_df["School Size"]=pd.cut(school_size_bins_df["Total Students"], size_bins, labels=labels)

# Formatting
school_size_bins_df["Total School Budget"] = school_size_bins_df["Total School Budget"].map("${:,.2f}".format)
school_size_bins_df["Per Student Budget"] = school_size_bins_df["Per Student Budget"].map("${:,.2f}".format)

school_size_bins_df
```

```python
# Calculate averages for the desired columns. 
size_math_scores = school_size_bins_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_size_bins_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_size_bins_df.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = school_size_bins_df.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = school_size_bins_df.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
school_summary_size_bins_df = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing
})

# Display results
school_summary_size_bins_df
```
